## 9. Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023

```python
from datetime import date,timedelta
d1=date(2023,1,1)
d2=date(2023,2,10)
delta=d2-d1
for i in range (delta.days+1):
  day=d1+timedelta(days=i)
  print(day)
```
